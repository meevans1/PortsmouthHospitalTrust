```python
# Start with loading all necessary libraries
import pandas as pd
import pyodbc
import re
import matplotlib.pyplot as plt
import numpy as np
```

```python
snomedct = pd.read_csv('sct2_Description_Snapshot-en_INT_20190731.txt',sep="\t",usecols=['effectiveTime','active','moduleId','conceptId','term'])
snomedct
```

```python
snomedct.dropna(inplace=True)
snomedct = snomedct[snomedct['active']==1]
snomedct = snomedct[snomedct['moduleId']==900000000000207008]
snomedct = snomedct[snomedct['term'].str.len()>2]
snomedct = snomedct[~snomedct['term'].str.isdigit()]
snomedct = snomedct[~snomedct['term'].str.contains('\^')]
snomedct = snomedct[~snomedct['term'].str.contains('\+')]
snomedct = snomedct[~snomedct['term'].str.contains('\[')]
snomedct = snomedct[snomedct['effectiveTime']>20170730]
snomedct
```

```python
# Python3 code to Check for  
# balanced parentheses in an expression 
open_list = ["[","{","("] 
close_list = ["]","}",")"] 
  
# Function to check parentheses 
def check(myStr): 
    open_stack = []
    close_stack = []
    for i in myStr: 
        if i in open_list: 
            open_stack.append(i) 
        elif i in close_list: 
            close_stack.append(i)
    if len(open_stack) == len(close_stack): 
        return False
    else: 
        return True
```

```python
fail = snomedct[np.vectorize(check)(snomedct['term'])].index
snomedct.drop(fail,inplace=True)
snomedct
```

```python
snomedct = snomedct.drop_duplicates('conceptId')
snomedct = snomedct.drop_duplicates('term')
snomedct
```

```python
snomedct = snomedct.drop(columns=['effectiveTime','active','moduleId','conceptId'])
```

```python
sql_conn = pyodbc.connect('DRIVER={SQL Server};'
                            'SERVER=L_AAGname;'
                            'DATABASE=name;'
                            'Trusted_Connection=yes') 
query = "set transaction isolation level read uncommitted select cn.Diagnosis,ps.SpecialtyDesc,pn.AmuTriage from tblClinicalNote cn inner join tblPatientSpell ps on ps.pkPatientSpellID=cn.fkPatientSpellID inner join tblPatientNote pn on pn.fkPatientSpellID=ps.pkPatientSpellID"
df = pd.read_sql(query, sql_conn)
df = df.dropna()
df
```

```python
df = df[df['Diagnosis']!='']
df = df[df['Diagnosis']!=' ']
#df = df[df.index<100]
```

```python
df_AMU = df[df['AmuTriage']!='-1']
```

```python
df_AMU['Diagnosis'] = df_AMU['Diagnosis'].str.replace(r'([0-9])\)',r'\1\.')
```

```python
df_AMU = df_AMU.drop(columns=['SpecialtyDesc','AmuTriage'])
```

```python
def findWholeWord(w):
    return re.compile(r'\b({0})\b'.format(w), flags=re.IGNORECASE).search
```

```python
def return_nan(tup):
    return(np.nan if len(tup[0]) == 0 else tup[0][0])


def get_indexes_match(l1, l2):
    #return([return_nan(np.where([x in e for x in l2])) for e in l1])
    return([return_nan(np.where([bool(findWholeWord(x)(e)) for x in l2])) for e in l1])


def merge(df1, df2, left_on, right_on):
    df1.loc[:, 'idx'] = get_indexes_match(df1[left_on].values,
                                          df2[right_on].values)
    df2.loc[:, 'idx'] = np.arange(len(df2))
    return(pd.merge(df1, df2, how='left', on='idx'))


merge(df_AMU, snomedct, left_on='Diagnosis', right_on='term')
```

```python
snomedct['term'] = snomedct['term'].str.lower()
df_AMU['Diagnosis'] = df_AMU['Diagnosis'].str.lower()
term_values = snomedct['term'].values
diagnosis_values = df_AMU['Diagnosis'].values
```

```python
mask = [any(i in words for i in term_values) \
        for words in diagnosis_values]

print(len(df_AMU[mask]))
```

```python
len(df_AMU[mask])/len(df_AMU)
```

```python
contain_snomed = 100*len(df_AMU[mask])/len(df_AMU)
not_contain_snomed = 100 - contain_snomed
plt.barh(1,contain_snomed,color='g')
plt.barh(1,not_contain_snomed,left=contain_snomed,color='r')
plt.text(contain_snomed/2,1,"{0:.0f}%".format(contain_snomed),ha='center')
plt.text(contain_snomed+not_contain_snomed/2,1,"{0:.0f}%".format(not_contain_snomed),ha='center')
plt.text(contain_snomed/2,0.4,"contain\nsnomed\nterm",ha='center')
plt.text(contain_snomed+not_contain_snomed/2,0.4,"don't\ncontain\nsnomed\nterm",ha='center')
plt.gca().axis('off')
```
