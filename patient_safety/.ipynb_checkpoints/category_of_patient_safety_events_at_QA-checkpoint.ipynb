```python
import pyodbc
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sql_conn = pyodbc.connect('Driver={SQL Server};'
                     'Server=L_AAGname;'
                     'Database=database_name;'
                     'Trusted_Connection=yes;')
```

```python
query = "select inc_dincident,inc_time,inc_locactual,inc_unit,inc_specialty,inc_loctype,inc_category from DatixCRM.dbo.incidents_main where inc_type='PAT' and inc_organisation='QA'"
```

```python
df = pd.read_sql(query,sql_conn)
```

```python
sizes = df['inc_category'].value_counts()

# the top 7
series2 = sizes[:7].copy()

# others
new_row = pd.Series(sizes[7:].sum(),index=['OTHER'])

# combine top 7 with others
series2 = pd.concat([series2,new_row])

explode = (0,0,0,0,0.1,0,0,0)
labels = series2.index

fig, ax = plt.subplots(figsize=(12, 6))
plt.pie(series2,explode=explode,labels=labels,autopct='%1.0f%%');
ax.set_title("Patient safety events at QA");
```
